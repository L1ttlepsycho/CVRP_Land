```python
import os
import matplotlib.pyplot as plt
import pickle 
import numpy as np

%matplotlib inline
```

```python
# read data from am-res

data_dir='data/am-res'

path_list=[]

for dir in sorted(os.listdir(data_dir)):
    dir=os.path.join(data_dir,dir)
    #print(dir)
    list=[os.path.join(dir, f) for f in sorted(os.listdir(dir))] \
        if os.path.isdir(dir) else [dir]
    path_list.append(list)

print(path_list)

diction_am={}# key: set name, value: (size, cost list, time list)

for sub_list in path_list:
    for path in sub_list:
        fpath,fname=os.path.split(path)
        sname=fname.split('-')[2]
        graph_size=int(fname.split('-')[3][1:])
        #print(sname,graph_size)
       
        raw=pickle.load(open(path, 'rb+'))[0] #(cost,route,time)

        t_list=[]
        c_list=[]
        for r in raw:
            c_list.append(r[0])
            t_list.append(r[2])

        data=(graph_size,c_list,t_list)

        if sname in diction_am.keys():
            diction_am[sname].append(data)
        else:
            diction_am[sname]=[]
            diction_am[sname].append(data)

print(diction_am)
print(diction_am.keys())
```

```python
# read data from hgs-res

data_dir='data/hgs-res'

path_list=[]

for dir in sorted(os.listdir(data_dir)):
    dir=os.path.join(data_dir,dir)
    #print(dir)
    list=[os.path.join(dir, f) for f in sorted(os.listdir(dir))] \
        if os.path.isdir(dir) else [dir]
    path_list.append(list)

print(path_list)

diction_hgs={}# key: set name, value: (size, cost list, time list)

for sub_list in path_list:
    for path in sub_list:
        if not path.endswith('.pkl'): continue
        fpath,fname=os.path.split(path)
        sname=fname.split('-')[2]
        #print(fname.split('-')[3].split('off')[0][1:])
        graph_size=int(fname.split('-')[3].split('off')[0][1:])
        #print(sname,graph_size)
       
        raw=pickle.load(open(path, 'rb+')) #(cost,route,time)

        t_list=[]
        c_list=[]
        for r in raw:
            c_list.append(r[0]/1000)
            t_list.append(r[2])

        data=(graph_size,c_list,t_list)

        if sname in diction_hgs.keys():
            diction_hgs[sname].append(data)
        else:
            diction_hgs[sname]=[]
            diction_hgs[sname].append(data)

print(diction_hgs)
print(diction_hgs.keys())
```

```python
# plot of heuristics
for set_name in diction_am.keys():
    if set_name=='tai': continue # jump set tai
    am_data_sizes=[]
    am_costs=[]
    am_times=[]

    hgs_data_sizes=[]
    hgs_costs=[]
    hgs_times=[]
    
    lkh_data_sizes=[]
    lkh_costs=[]
    lkh_times=[]

    or_data_sizes=[]
    or_costs=[]
    or_times=[]

    sorted_am_data=sorted(diction_am[set_name], key=lambda x: x[0])

    sorted_hgs_data=sorted(diction_hgs[set_name], key=lambda x: x[0])

    sorted_lkh_data=sorted(diction_lkh[set_name], key=lambda x: x[0])

    sorted_or_data=sorted(diction_or[set_name], key=lambda x: x[0])

    for data in sorted_am_data:    
        am_data_sizes.append(data[0])
        am_costs.append(np.mean(data[1]))
        am_times.append(np.mean(data[2]))

    for data in sorted_hgs_data:
        hgs_data_sizes.append(data[0])
        hgs_costs.append(np.mean(data[1]))
        hgs_times.append(np.mean(data[2]))

    for data in sorted_lkh_data:
        lkh_data_sizes.append(data[0])
        lkh_costs.append(np.mean(data[1]))
        lkh_times.append(np.mean(data[2]))

    for data in sorted_or_data:
        or_data_sizes.append(data[0])
        or_costs.append(np.mean(data[1]))
        or_times.append(np.mean(data[2]))

        
    #print(data_sizes,costs, times)
    plt.plot(lkh_data_sizes,lkh_costs,'-',label='LKH')
    plt.plot(hgs_data_sizes,hgs_costs,'-',label='HGS')
    plt.plot(or_data_sizes,or_costs,'-',label='Or-Tools')
    plt.xlabel('data size')
    plt.ylabel('costs')
    plt.title(f'Set {set_name} Cost(scaler=1000)' )
    plt.legend()
    plt.savefig(f'pics/cost-heu-set-{set_name}')
    plt.show()

    plt.plot(lkh_data_sizes,lkh_times,'-',label='LKH')
    plt.plot(hgs_data_sizes,hgs_times,'-',label='HGS')
    plt.plot(or_data_sizes,or_times,'-',label='Or-Tools')
    plt.xlabel('data size')
    plt.ylabel('times')
    plt.title(f'Set {set_name} Run Time(s)' )
    plt.legend()
    plt.savefig(f'pics/time-heu-set-{set_name}')
    plt.show()
    
```
